```python
##Pandas는 python으로 data 분석을 하기 위한 최적의 module

'''판다스의 데이터 타입'''

## Pandas는 새로 정의된 2가지 데이터 타입을 사용

## 1. Series: Numpy의 1차원 Array와 상당히 유사
#     1차원의 같은 데이터 타입이 저장될 수 있는 자료구조
#     Ndarray를 기반으로 해 Pandas에서 새롭게 정의한 자료구조이다
## 2. DataFrame
#     Series를 모아서 Table 형태로 구성한 자료구조 (2차원. R의 데이터 프레임과 동일)


'''판다스 준비'''

# [Step 1] Pandas module 설치
# 아나콘다에 설치 : conda install pandas 

import numpy as np
import pandas as pd


'''Series 생성( numpy array와 비교)'''

''' Numpy Array생성 방법 '''
# 리스트를 활용해 nparray 생성
arr = np.array([-1, 5, 8, 10], dtype="float64")

# 또는 
arr = np.array([-1, 5, 8, 10], dtype=np.float64)
print(arr.dtype)

# 서로 다른 데이터 타입을 넣는다면 ?
arr = np.array([-1, 5, 3.14, "Hello"])
print(arr)
print(arr.dtype)


arr = np.array([-1, 5, 3.14, "Hello"], dtype=np.object)
print(arr)
print(arr.dtype)
print(arr[2])
print(type(arr[2]))

print("=" * 30)


'''Series 생성'''
s= pd.Series([-1, 5, 3.14, "Hello"], dtype=np.object)
s= pd.Series([-1, 5, 3.14, "Hello"], dtype="object")
print(s)

print("=" * 30)

```

```python
# 1. Series의 Value부분 출력
print("Series의 value 부분 : {}".format(s.values))

# 2. Series의 data type(dtype)
print("Series의 type 부분 : {}".format(type(s.values)))
print("Series의 type 부분 : {}".format(s.values.dtype))
print("Series의 type 부분 : {}".format(s.dtype))


# 3. Series의 index 부분 출력
print("Series의 index 부분 : {}".format(s.index))
print("Series의 index 부분 : {}".format(type(s.index)))


# Series는 index를 따로 지정할 수 있다 
# 기본은 숫자 index
# 아래와 같이 index를 문자로 변경
s= pd.Series([-1, 5, 8, 10],
             dtype=np.int32,
            index = ["a","c","k","f"])
print(s)

# index를 새롭게 지정해도 숫자 인덱스를 사용가능
# - 기본 내장 숫자 인덱스가 있기 때문
print(s[1])
print(s["k"])


# - 찾으려는 index가 2개 이상 존재할 경우 series로 출력된다 
s= pd.Series([-1, 5, 8, 10],
             
             dtype=np.int32,
            index = ["k","c","k","f"])

print(s["k"])


```

```python
'''Series의 Indexing과 Slicing'''
# ndarray에서 indexing 및 slicing 사용 가능
# Series도 slicing 할 수 있다  -> series로 출력
# 새로운 index를 이용해 slicing을 할 때 범위에 조심해야

s = pd.Series([-1,5,8,10],
             dtype = np.int32,
             index =["a","c","k","f"])

# 범위 조심
print(s[0:2]) # end exclusive
print(s["a":"k"])   # end inclusive

# boolean indexing 가능
# 짝수만 뽑기
print(s[s%2==0]) 

# Fancy indexing도 가능
print(s[[0,2]])


'''집계함수 모두 이용 가능'''

#집계함수 사용가능
print(s.sum())
```

```python
'''python의 특이한 for문'''
'''리스트 생성에 이용'''

## for문
## for 문을 list만들 때 사용할 수 있다 
myList = [1,2,3,4]

## 생성하고 싶은 list는 각 요소에 *2를 한 list
## [2,4,6,8}
result = myList *2
print(result) # [1, 2, 3, 4, 1, 2, 3, 4]

# 기존의 for문을 이용하는 방법 -> 비효율적
result = list()
for i in myList:
    result.append(i*2)
print(result)


# python에서만 이용하는 축약된 for문 이용
    # 리스트 안에 넣으면서 만들 수 있는 방법
result = [x*2 for x in myList]
print(result)
    # 조건도 부여 가능
result = [x*2 for x in myList if x>2]
print(result)



# dict에도 사용 할 수 있을까?  => 가능
result = {"stu" + str(x) : x**2 for x in range(0,10)}
print(result)


```

```python
from datetime import date, datetime, timedelta



today = datetime. today()
print(today)

'''주, 일, 시, 분 초 차이 계산: timedelta'''

# timedelta: weeks, days, hours, minutes, seconds (연과 월은 불가능함)
day = timedelta(days=-2)
print(today + day)

hours= timedelta(hours=-9)
print(today + hours)

seconds= timedelta(seconds=-60)
print(today + seconds)

'''연과 월의 차이 계산 : relativedelta'''
# relativedelta: month, year 가능

from dateutil.relativedelta import relativedelta
 
years = relativedelta(years=-9)
print(today + years)

months= relativedelta(months=-8)
print(today + months)


'''오늘날짜 지정'''
# 오늘 날짜가 3월 31일 => 1달 전? 2월 28일, 2월 29일 
# 오늘 날짜의 지정이 필요. 아래 모듈과 함수 사용
from dateutil.parser import parse   # dateutil패키지 parser모듈 parse함수
                                        

today = parse("2019-03-31")
print("오늘의 날짜: {}".format(today))

# 3월 28~31일 까지는 모두 2월 28일로 출력됨

months= relativedelta(months=-1)
print(today + months)
```

```python
'''Exception Handling'''

## 의도치 않는 에러가 발생될 경우가 있다
## 프로그래밍 단계에서는 문제가 없지만 
## 실행시 문제가 발생할 수 있는 코드가 있다.

''' What is exception handling??'''

## 에러의 종류 중에, 복구가 가능한 경우를 exception error 라고 지칭함

## 대표적으로 0으로 나누는 명령어

# 10/0  # Exception 발생 =
      # 정지
      # 이를 극복해서 프로그램이 중지되지 않도록 하는 것이 
      # exception handling!

''' The way to practice exception handling'''
    
###### exception handling 구문#####

try:
    print(10/0)
    
except Exception as inst:
    print(inst) # 예외상황에 대한 내용을 출력

finally : 
    print("호호")

# finally 부분을 찍고 아래로 프로그램이 계속 진행된다
# 데이터 베이스 에러 발생을 우회하도록 할 때 사용


```

```python

Boolean indexing

import numpy as np
import pandas as pd

data = {
    "이름": ["홍길동","최길동","이지안","박동훈"],
    "학과": ["컴퓨터","철학","영어영문","의학"],
    "학년": [1,3,4,2],
    "학점": [3.1,2.4,4.5,3.3]
}

df = pd.DataFrame(data,
                 columns=["학과","학년","이름","학점","등급"],
                 index=["one","two","three","four"])
display(df)

# 행 : slicing만 가능 , 숫자 인덱스(인덱스)로 가능
# 열 : indexing만 가능, 숫자가 안돼 column명으로만



# row inexing 할 때 단일 행을 못 들고 옴
# fancy indexing이 지원되지 않아요, 완전 불편!

df.loc["two"]
# col이 index로 빠지고 해당 내용이 series로 나옴   # loc함수에 index이름 사용하면 단일 row 추출 가능
# df.loc[2]   # loc 함수에 숫자 index 사용하면 단일row 추출 불가능 
df.loc["one":"three"]  # loc와 index로 slicing 가능 

# 떨어져있는 몇개의 row 가져오기
# loc 이용하면 fancy indexing 가능, indexing 이름으로는 가능하나, 숫자indexing으로는 불가
df.loc[["one","three"]]
# df.loc[[0,3]]: loc는 숫자 indexing과 사용 안함!!



## row, col 모두 영역 잡아서 들고오기
# slicing: col또한 col로 slicing 가능 

display(df.loc["one":"three", "이름"])
display(df.loc["one":"three", "이름":"등급"])
display(df.loc["one":"three", ["이름","등급"]])



# cs에서 특정 대상에 대해 4가지 작업이 가능함
# 추가, 삭제, 수정, 추출
# C: create 추가
# R: read 검색/추출
# U: update 수정 
# D: delete 삭제

```

```python
import numpy as np
import pandas as pd

data = {
    "이름": ["홍길동","최길동","이지안","박동훈"],
    "학과": ["컴퓨터","철학","영어영문","의학"],
    "학년": [1,3,4,2],
    "학점": [3.1,2.4,4.5,3.3]
}

df = pd.DataFrame(data,
                 columns=["학과","학년","이름","학점","등급"],
                 index=["one","two","three","four"])
display(df)


# row 추가하려면?
# loc로 가능
df.loc["five"]=["화학",2,"김연아",4.5,np.nan]
            # 순서대로 입력
display(df)

df.loc["six","이름":"학점"]=["아이유",4.4]    # 부분만 추가
display(df)

df.loc["seven",["학과","이름"]]=["몰라요","누굴까"]     # 
display(df)

## 원래 존재하는 row에 대해서 assign작업을 하면 당연히 updatae작업이 수행되요!
df.loc["one","이름":"학점"] = ["김연아",4.5]    # 골라내서 추가

display(df)
## 삭제를 해보아요!
new_df = df.drop("학점", axis=1)   # 어떤 방향으로 뭘 지울건지
# axis=1   column=1
# inplace를 생략하면 원본 유지
# 삭제된 복사본을 return

display(new_df)

# df.drop("학점", axis=1,inplace=True)    # inplace=False    =>  default 
# display(df)     # 원본이 변경되고 return은 none

new_df=df.drop("one",axis=0)    # 0 : 행, 1: 열
display(new_df)

# index 중요해! => 내가 어떤 index를 골라서 할지

# new_df=df.drop("oneakak",axis=0)
# display(new_df)
# # Error 없는데 어떻게 지워요!

# 행을 삭제할 때 slicing으로 삭제하는 건 지원하지 않아요!
# new_df=df.drop("one":"three",axis=0)

# 행을 삭제할 때 fancy indexing은 지원해요!
new_df = df.drop(["one","three"],axis=0)
display(new_df)

# row indexing: loc[ ]
```

```python

import numpy as np
import pandas as pd

data = {
    "이름" : ["스폰지밥","뚱이","징징이","다람이"],
    "학과" : ["요리","바보","클라리넷","과학"],
    "학년" : [1,3,4,2],
    "학점" : [3.1, 2.1, 3.8, 4.5]
}

df = pd.DataFrame(data,
                 columns = ["학과","학년","이름","학점","등급"],       
                 index = ["one","two","three","four"])   
display(df)

# Boolean indexing은 mask를 사용해요!

# 학점이 3.0 이상인 것 뽑고싶어
df['학점'] >= 3.0     # [True, False, True, True]

# 학점이 3.0 이상인 사람의 학과와 이름을 출력하세요!
# df.loc[행,열]
# 행 부분에 mask 로 들어감

df.loc[df['학점'] >= 3.0,["학과","이름"]]
```

```python

import numpy as np
import pandas as pd

data = {
    "이름" : ["스폰지밥","뚱이","징징이","다람이","집게사장"],
    "학과" : ["요리","바보","클라리넷","우주과학","경제"],
    "학년" : [1,3,4,2,6],
    "학점" : [3.1, 2.1, 3.8, 4.5,4.2]
}

df = pd.DataFrame(data,
                 columns = ["학과","이름","학점","학년","등급"],       
                 index = ["one","two","three","four","five"])

display(df)


## 1. 이름이 "뚱이" 인 사람을 찾아 이름과 학점을 DataFrame으로 출력
pd.DataFrame(df.loc["two",["이름","학과"]])
display(df.loc[df["이름"]=="뚱이",["이름","학과"]])
## 2. 학점이 (3.1, 2.1)인 사람을 찾아 하고가, 이름 학점을 DataFrame으로 출력
# df.loc[[df['학점'] ==(3.1),df['학점'] ==(2.1)],["학과","이름"]]
display(df.loc[(df['학점'] >2.0) & (df['학점'] <3.2) ,["학과","이름"]])

# df.loc[df['학점'] ==(2.1),["학과","이름"]]
# df.loc["five"]=["화학",2,"김연아",4.5,np.nan]
## 3. 학점이 3.0을 초과하는 사람을 찾아 등급을 "A"로 설정하세요!
df.loc[df["학점"]>3.0, "등급"] = "A"
display(df)
```

```python
## DataFrame을 제어해보자
## Random seed를 잡아서 재현성을 확보 
## [0,10)정수형 난수를 균등분포로 (6,4)로 생성

import numpy as np
import pandas as pd

np.random.seed(10)
data = np.random.randint(0,10,(6,4))
print(data)

df= pd.DataFrame(data)

## Column과 index를 다시 정의해 보자
df.columns = ["A","B","C","D"]
df.index = pd.date_range("20190106",periods= 6) # "시작일".periods = 일수 (2019년 2월 26일 부터 6일 임을 명시)
                                                # "시작일", "종료일"

    
## NaN값(결치값)을 포함하는 새로운 컬럼 "E"를 추가해보자
df["E"] = [5,np.nan,4,np.nan,2,9]


'''결측값의 측정'''

## NaN은 결치값이지만 float로 간주됨 ****
 # 연간이 가능. 연산결과는 NaN

    
    
## Sample로 결치값이 포함된 DataFrame을 생성했다. 이를 이용해

'''결측값을 삭제/변형하는 방법'''

'''결측값의 삭제'''
# R에서는 제어문으로 삭제함
# python은 처리가 용이함

# 결측값을 포함한 행 지우기 
    # how: any(결측값이 하나라도 있으면 삭제 ), all(모든 값이 Nan일때 지우기)
    # inplace:원본처리 False(=default, 원본남기기), True(원본지우기 )

new_df = df.dropna(how="any", inplace =False)
    # how값을 text로 줘야함에 유의 

display(df)
display(new_df)

'''결측값의 치환(.fillna())'''
new_df= df.fillna(value=0,inplace=False)
display(new_df)


'''결측값 판별(.isnull())'''
## E column에 NaN이 포함된 행을 찾아서 B와 C column값을 출력해보자

df.loc[df.isnull()["E"],["B","C"]]

# loc 숫자 쓸수 없다 행 선택 함수
# iloc 숫자 사용 가능한 행 선택 함수
###########################################################

# 기본적인 DataFrame의 제어에 관한 내용(indexing, slicing, Fancy indexing, NaN 처리)
```
