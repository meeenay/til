```python
## 문제 6. 
## 앞에서부터 읽을때나 뒤에서 부터 읽을 때 모양이 같은 수를
## 대칭수라고 한다

## 예) 9009, 1111, 303

## 두 자리 수를 곱해서 만들 수 있는 대칭수 중에 
## 가장 큰 수는 9009 (91*99)입니다

## 세 자리 수를 곱해서 만들 수 있는 가장 큰 대칭수는 얼마? 

'''
100*100 = ? => 대칭수인지 확인=> 저장
100*101 = ? => 대칭수인지 확인 => 저장
100*102 = ? => 대칭수인지 확인 => 저장 

대칭수인지 확인하는 것을 함수로 만들면 좋을 것 '''


def check_number(x): 
    str1 = str(x)
    tmp = list(str(x))
    tmp.reverse()
    if str1 == "".join(tmp):
        return True
    else:
        return False
    
print(check_number(1221))


# 아래 솔루션은 알고리즘의 효율이 좋지 않음 
# 100*101 , 101*100 두번씩 하는 것이므로

def calcul_number(x): # x: 1이면 2자리수끼리 곱
                   # x: 2이면 3자리수끼리 곱
    result = []
    for i in range(10**x,10**(x+1)):
        for j in range(10**x,10**(x+1)): 
            if check_number(i*j):
                result.append(i*j)
    print(max(result))

calcul_number(2)   

# 이대로 두면 하드 코딩
```
